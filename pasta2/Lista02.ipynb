func inverterNumero(num int) int {
	strNum := strconv.Itoa(num)
	invertido, _ := strconv.Atoi(reverseString(strNum))
	return invertido
}

func reverseString(s string) string {
	runes := []rune(s)
	for i, j := 0, len(runes)-1; i < j; i, j = i+1, j-1 {
		runes[i], runes[j] = runes[j], runes[i]
	}
	return string(runes)
}

func calcularMMC(a, b int) int {
	return a * b / calcularMDC(a, b)
}

func calcularMDC(a, b int) int {
	for b != 0 {
		a, b = b, a%b
	}
	return a
}

func contaDigitos(a, b, c int) int {
	max := a
	if b > max {
		max = b
	}
	if c > max {
		max = c
	}

	digitos := 1
	for max >= 10 {
		max /= 10
		digitos++
	}
	return digitos
}
